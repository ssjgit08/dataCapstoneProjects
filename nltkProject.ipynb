## Text Preprocessing

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

#### Load Data 

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

#### Lowercase

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

#### Stopword Removal

```python
en_stopwords = stopwords.words('english')
```

```python
en_stopwords.remove('not')
```

```python
data['review_no_stop_words'] = data['review_lowercase'].apply(lambda x: ' ' .join([ word for word in x.split() if word not in (en_stopwords)]))
```

```python
data.head()
```

```python
data['review_no_stop_words'][0]
```

#### Punctuation 
We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word star. We want to keep this as it might add meaning to the review.

```python
data['review_no_stop_words_no_punct'] = data.apply(lambda x: re.sub(r"[*]", "star", x['review_no_stop_words']), axis = 1)
```

```python
data.head()
```

```python
data['review_no_stop_words_no_punct'] = data.apply(lambda x: re.sub(r"[^\w,\s]", "", x['review_no_stop_words_no_punct']), axis = 1)
```

```python
data.head()
```

#### Tokenizing

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stop_words_no_punct']), axis = 1 )
```

```python
data['tokenized'][0]
```

#### Stemming

```python
ps = PorterStemmer()
```

```python
data['stemmed'] = data['tokenized'].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

#### Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
data['lemmatized'] = data['tokenized'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
data['lemmatized'][0]
```

```python
data.head()
```

#### N-grams

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2)).value_counts())
print(bigrams)
```
